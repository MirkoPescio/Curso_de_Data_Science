```python
# La particularidad es que de esta forma se asigna de forma constante para todas las filas
# Por lo tanto hay que buscar otra forma de emular otro comportamiento
```

```python
# Agregar una nueva columna con un array
```

```python
# importar numpy
import numpy as np
```

```python
# Crear un array de 1000 elementos
# Porque este dataframe tiene 1000 filas
np.arange(0, 1000)
```

```python
len(np.arange(0, 1000))
```

```python
# Efectivamente son 1000 elementos los que tiene el array
```

```python
# Crear un array de 1000 elementos
puntaje_lenguaje = np.arange(0, 1000)
```

```python
# Tamaño del array
len(puntaje_lenguaje)
```

```python
# agregar una nueva columna al dataframe con un array
df_examenes["languaje score"] = puntaje_lenguaje
df_examenes
```

```python
# Ahora no tenemos valores constantes para la nueva columna, sino, valores secuenciales
# Sin embargo, para esta columna buscamos valores que tengan otro comportamiento
# no uno secuencial de forma ascendente, sino, uno aleatorio o random
```

```python
# Crear números enteros aleatorios entre 1 y 100
np.random.randint(1, 100+1, size=1000)
# primer argumento => valor de inicio de los números aleatorios (inclusivo)
# segundo argumento => valor límite de los números aleatorios (exclusivo)
# tercer argumento => tamaño del array
```

```python
# Nos devuelve un array de 1000 elementos con valores aleatorios desde 1 hasta 100
# Verificamos
len(np.random.randint(1, 100+1, size=1000))
```

```python
# Creo la lista de 1000 elementos con valores aleatorios
puntaje_lenguaje_random = np.random.randint(1, 100+1, size=1000)
```

```python
# Agrego la nueva columna 'languaje_score' al dataframe con el array de valores aleatorios
df_examenes["languaje_score"] = puntaje_lenguaje_random
df_examenes
```

```python
# Crear números aleatorios entre 1 y 100
np.random.uniform(1, 100+1, size=1000)
```
